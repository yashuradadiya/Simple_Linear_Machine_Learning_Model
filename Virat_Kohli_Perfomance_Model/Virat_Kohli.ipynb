### Q-7.  Create pie chart of above analyzed data 

```python
total_value = {}
for i in range(1,8):
    total_value["Betting Position "+str(i)] = (df['Pos']==i).sum()
total_value
```

```python
plt.figure(figsize=(6,6))
plt.pie(total_value.values(),labels=total_value.values(),colors = ["#7E8BCD", "#91CCB6", "#D892C0", "#7FD3E5", "#A086BC", "#F9A881", "#C6E0B4"])
plt.legend(loc='best', labels=total_value.keys(),bbox_to_anchor=(1,0.4))
plt.title('Count batting positions played at paticular position')
plt.show()
```

### Q-8. look at the total runs scored by Virat Kohli in different positions:
