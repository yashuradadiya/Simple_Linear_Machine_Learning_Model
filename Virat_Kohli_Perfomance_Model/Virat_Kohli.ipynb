### Q-9. find out the number of centuries scored by Virat Kohli while batting in the first innings and second innings:

### Q-11. Find out against which team Virat Kohli scored most of his runs:

```python
df['Opposition'].unique()
```

```python
df.groupby('Opposition')['Runs'].sum()
```

```python
plt.figure(figsize=(15,5))
plt.bar(sorted(df['Opposition'].unique()),df.groupby('Opposition')['Runs'].sum())
plt.xticks(rotation=90)
contries = sorted(df['Opposition'].unique())
runs_c = df.groupby('Opposition')['Runs'].sum()
for i in range(len(contries)):
    plt.text(contries[i], runs_c[i], str(runs_c[i]), ha="center", va="bottom")
```

against **Sri Lanka** team Virat Kohli scored most of his runs

### Q-12. Find out against which team Virat Kohli scored most of his centuries:

```python
count_centuries = df.groupby('Opposition')['Runs'].agg(lambda x: (x >= 100).sum())
count_centuries
```

```python
plt.figure(figsize=(15,5))
plt.bar(sorted(df['Opposition'].unique()),count_centuries)
plt.xticks(rotation=90)
Oppositions =sorted(df['Opposition'].unique())
con_o = count_centuries
for i in range(len(Oppositions)):
    plt.text(Oppositions[i], con_o[i], str(con_o[i]), ha="center", va="bottom")
plt.show()
```

against **Australia** team Virat Kohli scored most of his centuries
